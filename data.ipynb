```python
import numpy as np
```

```python
from neurogym.envs.delaymatchsample import DelayMatchSampleDistractor1D
```

```python
X = []
Y = []
env = DelayMatchSampleDistractor1D()
N = 5
for _ in range(N):
    env.new_trial()
    # input sequence
    x_trial = env.ob      # shape: (T, 33)

    # correct labels
    y_trial = env.gt      # shape: (T,)
    
    X.append(x_trial)
    Y.append(y_trial)

X = np.stack(X)           # shape: (N, T, 33)
Y = np.stack(Y)           # shape: (N, T)
```
